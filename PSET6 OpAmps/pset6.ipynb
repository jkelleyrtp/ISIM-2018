### Op-amp voltage follower
-------

#### Does the follower work as advertised?
-Yes, it follows in the input wave with a lag period


#### At what frequency apprxomiately do you notice a significant difference between output and input?
- We see the frequency start to peel apart at 50 kHz and be noticeable at 250 kHz

```python
lagging = pd.read_csv("data/follower_lagging_slower.csv", skiprows = 9)
plt.figure(dpi = 300)
plt.scatter(lagging['Time (s)']+4.096e-05, lagging["Channel 1 (V)"], s = 1, label = "Input Voltage")
plt.scatter(lagging['Time (s)']+4.096e-05, lagging["Channel 2 (V)"], s = 1, label = "OpAmp Output Voltage")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.xlim(0, 8.096e-06)
plt.title("OpAmp response to 250 kHz wave")
```

#### How long does it take for the output to catch up to the input for a square wave?
- We see the lag time to be about 2 µs

```python
square = pd.read_csv("data/square_time_difference.csv", skiprows = 9)
plt.figure(dpi = 300)
plt.scatter(square['Time (s)']+0.00512, square["Channel 1 (V)"], s = 1, label = "Input Voltage")
plt.scatter(square['Time (s)']+0.00512, square["Channel 2 (V)"], s = 1, label = "OpAmp Output Voltage")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.xlim(0.00045, 0.00055)
plt.title("OpAmp response to 250 kHz wave")
```

#### Set the frequency back to 1 kHz, amplitude 1V, centered at 0V. Why doesn't this circuit work?
- It dips below 0 volts

```python
square = pd.read_csv("data/time_difference_cutoff.csv", skiprows = 9)
plt.figure(dpi = 300)
plt.scatter(square['Time (s)'] + abs(square['Time (s)'][0]), square["Channel 1 (V)"], s = 1, label = "Input Voltage")
plt.scatter(square['Time (s)'] + abs(square['Time (s)'][0]), square["Channel 2 (V)"], s = 1, label = "OpAmp Output Voltage")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
#plt.xlim(0.00045, 0.00055)
plt.title("Nonfollower circuit")
```
