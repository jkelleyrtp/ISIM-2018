```python
import warnings; warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats

%matplotlib inline
```

### Plot 1: RC Circuit Voltage

```python
file = "DCCapacitorCircuit.csv"
data = pd.read_csv(file, skiprows = 8)
data["Time (s)"] = data["Time (s)"] + 5
data = data.iloc[3500:]


R = 100000
C = 10e-6


# Generate the figure
data_time = data["Time (s)"]
data_voltages = data["Channel 1 (V)"]

plt.figure(dpi = 180)
plt.scatter(data_time, data_voltages, s=1, label = "Voltage over RC circuit with time constant of "+str(R*C)+" s" )
plt.title("Voltage (V) vs Time (s) for RC Circuit on Potentiometer")
plt.xlabel("Time(s)")
plt.ylabel("Voltage (V)")
plt.legend()
```

### Experimental vs Analytical Solution

```python
t0 = data.iloc[0]['Time (s)'] + 5
v0 = data.iloc[0]['Channel 1 (V)']

times = np.linspace(0, 10, 4499)
pred_voltages = v0 * np.power(np.e, -1 * times /(R*C) )

plt.figure(dpi = 180)
plt.scatter(data_time - 4.375, data_voltages, s=.5, label = "Collected Data")
plt.scatter(times, pred_voltages, s=.5, label = "Analytical Solution")

plt.title("Voltage vs Time compared for Theroetical and Experimental RC Circuit")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")

plt.legend()
```

```python
plt.figure(dpi = 180)
I = pred_voltages/R *1e6
plt.scatter(times, I, s=.5, label = "Analytical Solution")

plt.title("Current vs Time for Experimental RC Circuit")
plt.xlabel("Time (s)")
plt.ylabel("Current (microamps)")
plt.ylim(0, np.max(I))

plt.legend()
```

```python
plt.figure(dpi = 180)
plt.plot(C * np.abs(np.diff(pred_voltages)/0.00125) * 1e6)

plt.title("Current vs Time for Experimental RC Circuit")
plt.xlabel("Time (ms)")
plt.ylabel("Current (microamps)")
```

The data shows us that the current through the capacitor is equal to the capacitance times the change in voltage with respect to time.
