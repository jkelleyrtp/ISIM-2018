#### The code that generated these plots is shown below.

```python
def plot_fig_1():
    data = pd.read_csv("BodePlot.csv", skiprows = 20)
    w = data['Frequency (Hz)']
    magnitude = data['Channel 2 Magnitude (dB)']
    phase = data['Channel 2 Phase (°)']
    
    fig = plt.figure(dpi = 150)
    ax1 = plt.subplot(211)
    plt.semilogx(w, magnitude)    # Bode magnitude plot
    plt.title("Bode Plot for RC circuit")
    plt.ylabel("Magnitude (dB)")
    
    ax2 = plt.subplot(212)
    plt.semilogx(w, phase)  # Bode phase plot
    plt.ylabel("Phase (°)")    
    plt.subplots_adjust(wspace=0, hspace=.05)
    plt.xlabel("Frequency (Hz)")
    plt.show()
```

```python
def plot_fig_2():
    data = pd.read_csv("BodePlot2.csv", skiprows = 20)
    w = data['Frequency (Hz)']
    magnitude = data['Channel 2 Magnitude (dB)']
    phase = data['Channel 2 Phase (°)']
    
    fig = plt.figure(dpi = 150)
    ax1 = plt.subplot(211)
    plt.semilogx(w, magnitude)    # Bode magnitude plot
    plt.title("Bode Plot for CR circuit")
    plt.ylabel("Magnitude (dB)")
    
    ax2 = plt.subplot(212)
    plt.semilogx(w, phase)  # Bode phase plot
    plt.ylabel("Phase (°)")    
    plt.subplots_adjust(wspace=0, hspace=.05)
    plt.xlabel("Frequency (Hz)")
    plt.show()
```
