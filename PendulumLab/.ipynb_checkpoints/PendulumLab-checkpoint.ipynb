```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats


%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

### First, let's load the calibration data 

```python
calibrated_data = np.array([[0, 800],
                            [20, 725],
                            [40, 671],
                            [60, 617],
                            [80, 551],
                            [90, 536],
                            [100, 510],
                            [120, 450],
                            [140, 392],
                            [160, 330],
                            [180, 275] ])
```

```python
### Graph the data too
angles = calibrated_data[:,0]
voltages = calibrated_data[:,1]

plt.figure(dpi = 400)
plt.scatter(angles, voltages)
plt.title("Voltage (mV) vs Pendulum Angle (°) ")
plt.xlabel("Pendulum Angle (°)")
plt.ylabel("Voltage (mV)")


# Seaborn overwrites previous plots so we'll just add linear guess now
sb.regplot(angles, voltages)
```

The data seems quite linear! Let's create a line to match the voltage to the pendulum angle.

Remember: The voltage is regulated from 0V to 1V and is expected to roll around at 1V to 0V

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(voltages, angles )

def angle_from_voltage(voltage):
    return voltage * slope + intercept

# Now we can use our function to graph the angle from the potentiometer voltage
```

#### Let's plot the raw data collected from the potentiometer

```python
# Load the Wavegen Data
file = "PotentiometerPendulum.csv"
data = pd.read_csv(file, skiprows = 4, skipfooter = 500)
```

```python
# Generate the figure

data_time = data["Time (s)"]+ 25
data_voltages = data["Channel 1 (V)"] * 1000

plt.figure(dpi = 400)
plt.scatter(data_time, data_voltages, s=1)
plt.title("Voltage (mV) vs Time (s) for Pendulum on Potentiometer")
plt.xlabel("Time(s)")
plt.ylabel("Voltage (mV)")
```

```python
#### Let's convert the raw voltage to angle

data_angles = angle_from_voltage(data_voltages)
```

```python
plt.figure(dpi = 400)

plt.scatter(data_time, data_angles, s=1)
plt.title("Angle (°) vs Time (s) for Pendulum on Potentiometer")
plt.xlabel("Time(s)")
plt.ylabel("Angles (°)")
```
