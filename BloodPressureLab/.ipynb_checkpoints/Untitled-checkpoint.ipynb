I chose a lower frequency cutoff of 200 mHz and a higher frequency cutoff of 2 Hz. 

Our 200 mHz cutoff required a high pass filter with an RC constant of 0.80 and the 2 Hz cutoff required a low pass filter with an RC constant of 0.4.

My final circuit had a 20k ohm and 2 x 1uf filter for the high pass and a 4.02k ohm and 2 x 100uf for the low pass.

We are using the 2.5 V as reference because the AD623 is bound from 0 to 5 volts, so the signal can be positive and negative without crossing the boundaries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pressure = pd.read_csv("pressure.csv")
plt.figure(dpi = 200)
plt.plot(pressure['0.73375266520268101'])
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (V)")
plt.title("Raw pressure data")
```

```python
pulse = pd.read_csv("pulse_full.csv", skiprows = 9)
plt.figure(dpi = 200)
plt.plot(pulse['Time (s)'], pulse['Channel 2 (V)'])
plt.title("Filtered pressure data")
plt.ylabel("Amplitude (Volts)")
plt.xlabel("Time (s)")
```

```python
y = np.fft.fft(pulse['Channel 2 (V)'])
freq = np.fft.fftfreq(len(pulse['Channel 2 (V)']), pulse['Time (s)'][1] - pulse['Time (s)'][0])

plt.figure(dpi = 200)
plt.xlim((0,10))
plt.ylim((0, 600))
plt.plot(freq, np.abs(y))
plt.plot(freq, np.angle(y))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Ampltiude")
plt.title("Frequency Analysis of Filtered Pressure Signal")
```

```python
def bode_plot():
    data = pd.read_csv("bode.csv", skiprows = 20)
    w = data['Frequency (Hz)']
    magnitude = data['Channel 2 Magnitude (dB)']
    phase = data['Channel 2 Phase (°)']
    
    fig = plt.figure(dpi = 150)
    ax1 = plt.subplot(211)
    plt.semilogx(w, magnitude)    # Bode magnitude plot
    plt.title("Bode Plot for complete blood pressure circuit")
    plt.ylabel("Magnitude (dB)")
    
    ax2 = plt.subplot(212)
    plt.semilogx(w, phase)  # Bode phase plot
    plt.ylabel("Phase (°)")    
    plt.subplots_adjust(wspace=0, hspace=.05)
    plt.xlabel("Frequency (Hz)")
    plt.show()
```
